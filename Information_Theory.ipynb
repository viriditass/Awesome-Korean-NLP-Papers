# **정보이론(Information Theory)**

정보이론이란, **어떤 메시지가 가지고 있는 정보량을 표현하는 방법**을 연구하는 것이다.

이렇게 어떤 메시지가 가지고 있는 정보량을 표현하기 위해선
1. **정보량이 정의되어야 한다.**
2. **정보를 표현하는 방법에 대해 논의가 필요하다.**

위와 같이 2가지의 조건이 필요하다.

## **Entropy**

Entropy란, **정보를 표현하는데 있어서 필요한 최소 자원량**이다. 다시말해 **정보를 인코딩하는데 있어서 필요한 최소 bit수**이다. 위에서 정보량에 대해 설명했던 예시를 모두 해당 정보를 표현하기 위한 최소 정보량을 알기 위해 UpDown 게임을 진행한 것이다.

이번에도 예시를 들기 위해 A와 B라는 사람이 있다 치고, 각자 자기 먹을 음식을 bit로 보낸다고 하자.


이때, A는
- 치킨를 먹을 확률이 0.25
- 피자를 먹을 확률이 0.25
- 햄버거를 먹을 확률이 0.25
- 밥을 먹을 확률이 0.25

이다.

반면, B는
- 치킨을 먹을 확률이 0.5
- 피자를 먹을 확률이 0.125
- 햄버거를 먹을 확률이 0.125
- 밥을 먹을 확률이 0.25

이다.

image.png

A의 경우, 위의 예시들을 봤으면 예상했듯이 최소 질문의 개수는 2개이다.

image.png

하지만, B의 경우는 다르다. B의 경우는 굳이 (치킨, 피자)중에 있니? (햄버거, 밥)중에 있니?라고 묻는 전략은 비효율적이다. 왜냐하면 이미 치킨이 50% 확률로 나타나기 때문이다. 그러므로 처음엔 (햄버거)를 먹을래 혹은 (나머지 셋중에 먹을래)라고 물어보는 것이 옳다. 그리고 나서 밥이 나머지 0.5의 확률중에 0.25를 차지함으로 밥을 먹을래 혹은 피자/햄버거를 먹을래 라고 물어보는 것이 옳다.

치킨, 피자, 햄버거, 밥을 각각 A, B, C, D로 표현하여 식을 적으면,

image.png

위의 식처럼, 평균적으로 1.75개의 질문을 하면 B가 어떤 음식을 먹을지 알아낼 수 있다. 만일, A와 B로 부터 각각 100개의 식사를 맞히기 위해서는 A한테는 200번의 질문을, B한테는 175번의 질문을 해야한다. 그러므로 **B가 A보다 더 적은 정보량을 생산**한다고 볼 수 있다.

왜냐하면 불확실성(랜덤성)이 적기 때문이다. 이를 식으로 정리한 것이 바로 **Claude Shannon**이다. Shannon은 이 불확실성의 측정을 **Entropy**라고 정의하였고, 이를 H라고 표시했으며, 단위는 bit로 정했다.

image.png

다시 한번 생각해보면, 질문 횟수는 A, B, C, D(치킨, 피자, 햄버거, 밥)의 발생 확률과 연관이 있다.

- 치킨: 0.5 => 1/2
- 피자: 0.125 => 1/8
- 햄버거: 0.125 => 1/8
- 밥: 0.25 => 1/4

image.png

그러므로 앞서 말한 '가능한 결과의 수'는 그 사건 발생확률의 역수이다.

image.png

위의 식이 아까 말한 Shannon의 식이다. 이 식은 **이산확률분포**일 때의 이야기이다. 이에 따라 **시그마**가 들어온다. 일어날 수 있는 각 사건에 대해 값을 구하고 다 더한다. 즉, 이 식에서 나온 **H가 해당 정보를 표현하기 위한 minimum값**이라는 것이다.

참고로 Entropy는 모든 사건이 같은 확률일때, 즉 확률 분푸가 uniform distribution일때 maximum값을 갖는다.

## **Cross Entropy**


Cross Entropy란, **어떤 문제에 대해 특정 전략을 쓸 때 예상되는 기댓값**이다.

image.png

식을 보면, 단번에 이해할 수 있을 것이다. 그래도 한번 또 예시를 들어, Entropy에서 설명을 했던 A와 B를 다시 불러오겠다. 그리고 이번엔 A에서 사용한 기법을 B에서 사용해보겠다.

image.png

계산을 해보면, 2 x 0.5) + (2 x 0.125) + (2 x 0.125) + (2 x 0.25) = 2 x 1 = 2이므로, 기존 1.75였던 entropy보다 0.25가 더 커진걸 알 수 있다. 정리하자면, 어떤 문제 p에 대해 확률분포로 된 어떤 전략 q를 사용할 때의 기댓값이 바로 cross entropy다.

다시, 위의 식에 예시를 대입하면 p = [0.5, 0.125, 0.125, 0.25]이고, q = [0.25, 0.25, 0.25, 0.25]가 되는 것이다. 따라서 q(i)를 학습하고 있는 상태라면 p(i)에 가까워질수록 cross entropy 값은 작아진다. **이런 특성때문에 cross entropy를 머신러닝에 많이 사용**하는 것이다.

image.png

이산형이 아니라 연속형인 확률본포에서는 시그마가 아니라 integral이 들어오게 된다. 바로 위의 식과 같은 형태이다.

### **Binary classfication과 Cross Entropy**

image.png

위 식은 logistic regression에서 보는 loss function이다. 이 식은 사실 정학하게는 cross entropy 식이다. 식을 하나씩 듣어보면, Binary classification에서는 0 또는 1로 두가지 class를 구분한다. 이를 수식으로 표현하면 y∈ { 0, 1 }이다. 우리가 어떤 대상이 1이라고 predict하는 확률 q(y) = 1을 y^로 놓겠다. 그러면 어떤 대상을 0으로 predict하는 확률 q(y) = 0은 (1 - y^)가 된다.

image.png

결국 위의 식처럼 정리가 된다. 따라서 logistic regression의 cost function은 단순히 우리가 봤던 cross entropy의 시그마를 풀어쓴 것에 불과했다.

### **Cross Entropy와 Log Loss**

cross entropy는 log loss로 불리기도 한다. 그 이유는 cross entropy를 최소화하는 것은 log likelihood를 최대화하는 것과 같기 때문이다.

image.png

어떤 데이터가 0 또는 1로 predict될 확률은 각각 y^, 1 - y^이므로 그 데이터의 likelihood 식을 위와 같이 세워볼 수 있다. (y가 0 또는 1의 값만 가질때 가능하다.)

image.png

y = 1일때 y^를 최대화시켜야 하고, y = 0일 때는 (1 - y^)를 최대화시켜야 한다. 여기에 log를 씌우면, 지수들이 내려와서 위와 같은 식이 된다.

최소화 해야하는 식은 우리가 아까 살펴본 cross entropy와 같은 식이다. 이런 이유로 cross entropy를 log loss라고 부른다. 왜냐하면 cross entropy 값이 커지면 log likelihood가 작아지기 때문에 cross entropy값을 작게 해야한다. 또 나아가 **negative log likelihood**라고도 표현한다.

## **KL divergence**

KL divergence는 정확히는 **Kullback-Leibler divergence(KLD)**로 두 확률분포의 차이를 계산하는데 사용하는 함수다. 어떤 이상적인 분포에 대해 그 분포를 근사하는 다른 분포를 사용해 샘플링을 한다면 발생할 수 있는 **정보 엔트로피** 차이를 계산한다. 또한, **상대 엔트로피(relative entropy)**, **정보 획득량(information gain)**, **인포메이션 다이버전스(information divergence)**라고도 한다.

위의 정의를 보면, 바로 이해가 안갈수 있다. 쉽게 설명하자면 그냥 Cross Entropy - Entropy라고 생각하면 된다. 즉 다시말해, KL divergence는 p와 q의 cross entropy에서 p의 entropy를 뺀 것이다. 결과적으론 두 분포의 차이를 나타낸다.

image.png

이를 식으로 표현하면 위와 같다. 보통 DKL(p|q) 혹은 KL(p|q)로 많이 표현한다.

image.png

cross entropy를 minimize 하는 것은, 어차피 H(p)는 고정된 상수값이기 때문에 결과적으로는 KL-divergence를 minimize 하는 것과 같다.

### **KL divergence의 특성**

image.png

위의 특징이 바로 KL divergence의 대표적인 특징 2개이다. "KL-divergence는 거리 개념이 아니다"라는 말로 매우 자주 등장하곤 한다.

1. **0이상이다.**

이는 직관적으로 바라보면 매우 당연한 소리다. entropy가 정보량의 minimum값이기 때문이다. KL divergence의 최소값이 0일땐, cross entropy와 entropy의 값과 동일할때 이다.

image.png

하지만, 보통 KL divergence가 0 이상인 것을 증명할 때는 **젠슨 부등식(Jensen's Inequality)**을 사용한다. 이 부등식을 사용하려면 convex function(아래로 볼록한 함수)를 먼저 엄밀하게 정의한다.

두 점 x1, x2와 그 사이에 있는 점 tx1 + (1 - t)x2를 보겠다.(t ∈ [0,1]). 가운데 점의 식은 두 점 x1, x2의 weighted sum이다. 그리고 이제 그 가운데 저을 함수 f(x)에 넣었을 때 나온 값 f(t(x1) + (1-t)x2)와 두 점 x1, x2를 각각 f(x)에 넣은 값의 weighted sum인 t(f(x1)) + (1-t)f(x2)를 비교한다. 

image.png

이러면 위와 같은 식이 나오게 된다.

image.png

이를 확률론의 맥락에서는 X가 random variable이고. f(·)가 convex function일 때, 위와 위의 식과 같이 표현한다.

image.png

위의 식처럼 KL-divergence에 관련해서 증명을 해보면. 우리의 KL-divergence 식에 있는 −log 를 f(x)로 두고 가운데의 weighted sum 또는 E[X]를 q(i) / p(i)로 두면 Jensen’s 부등식에 넣을 수 있다.

2. **거리 개념이 아니다.**


image.png

위에서 말했듯이 KL divergence하면 따라오는 말이 바로 "KL divergence는 거리 개념이 아니다"이다. 또한 대게 뒤따라오는 말은 "KL-divergence는 asymmetric하다"이다. 어떤 부분이 비대칭적인가? 바로 KL divergence에서 p와 q를 바꾼 값과 원래의 값이 다르다는 점이 비대칭적이다. 대칭적이라면 당연히 결과값이 같아야 한다. 위의 KL-divergence의 식을 살펴보면 만약에 두 확률분포 사이의 거리라면 **p에서 q 사이의 거리**와 **q에서 p 사이의 거리**가 같아야 한다. 하지만 KL-divergence는 그렇지 않다. 이러한 이유 덕분에 KL divergence가 distance metric이 아니라고 하는 것이다. 하지만 **거리 개념처럼 쓸 수 있는 방법**이 존재하는데 이게 바로 **Jensen-Shannon divergence**이다.
